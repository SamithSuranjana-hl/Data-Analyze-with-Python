# Finding is there any significant relationship between the columns ,

```python
corr = sales.corr(numeric_only=True)         #The argument suggests to pick onlu numerical columns
corr
```

```python
range(len(corr.columns))
```

```python
fig = plt.figure(figsize=(8,8))
plt.matshow(corr,cmap='BrBG')
plt.xticks(range(len(corr.columns)),corr.columns,rotation='vertical')
plt.yticks(range(len(corr.columns)),corr.columns)
```

```python
sales.plot(kind='scatter',x='Unit_Cost',y='Unit_Price',figsize=(10,6))
```

```python
sales.plot(kind='scatter',x='Customer_Age',y='Revenue',figsize=(8,6))
```

```python
ax = sales[['Profit','Age_Group']].boxplot(by='Age_Group',figsize=(10,6))
ax.set_ylabel('Profit')
```

```python
box_cols = ['Year',	'Customer_Age',	'Order_Quantity',	'Unit_Cost',	'Unit_Price',	'Profit'	]
ax=sales[box_cols].boxplot(figsize=(10,6))
```

```python
boxplot_cols = ['Year', 'Customer_Age', 'Order_Quantity', 'Unit_Cost', 'Unit_Price', 'Profit']

sales[boxplot_cols].plot(kind='box', subplots=True, layout=(2,3), figsize=(14,8))
```

# Column Wrangling
creating new columns or modifying existing ones.

```python
sales['Revenue_Per_Age'] = sales['Revenue']/sales['Customer_Age']
sales['Revenue_Per_Age'].head()
```

```python
sales['Revenue_Per_Age'].plot(kind='density',figsize=(14,6))
```

```python
sales['Revenue_Per_Age'].plot(kind='hist',figsize=(14,6))
```

Adding a new column called 'Calculated_Cost'

```python
sales['Calculated_Cost']=sales['Order_Quantity']*sales['Unit_Cost']
sales['Calculated_Cost'].head()
```

```python
(sales['Calculated_Cost'] != sales['Cost']).sum()
```

```python
sales.plot(kind='scatter',x='Calculated_Cost',y='Profit',figsize=(10,6))
```

```python
sales['Calculated_Revenue'] = sales['Order_Quantity']*sales['Unit_Price']
sales['Calculated_Revenue'].head()
```

```python
sales['Re'] = sales['Cost']+sales['Profit']
sales['Re'].head()
```
